```python
import pandas as pd

# Create sample data
data = {
    'Hours_Studied': [1,2,3,4,5,6,7,8],
    'Attendance': [50,60,65,70,75,80,85,90],
    'Result': [0,0,0,1,1,1,1,1]  # 0 = Fail, 1 = Pass
}

df = pd.DataFrame(data)

# Split data
X = df[['Hours_Studied', 'Attendance']]
y = df['Result']

# Train model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

# Predict
prediction = model.predict([[5, 75]])
print("Prediction (1=Pass, 0=Fail):", prediction)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```

```python
print(model.predict([[2, 60]]))  # Low → Fail
print(model.predict([[7, 85]]))  # High → Pass
```
